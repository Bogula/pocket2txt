```python
import requests
from bs4 import BeautifulSoup
```

```python
from readability.readability import Document

def extract_reader_text(urls):
    headers = {"User-Agent": "Mozilla/5.0"}
    results = {}

    for url in urls:
        try:
            response = requests.get(url, headers=headers, timeout=10)
            response.raise_for_status()

            doc = Document(response.text)
            reader_html = doc.summary()  # Main content HTML
            soup = BeautifulSoup(reader_html, 'html.parser')

            # Optionally extract title:
            title = doc.title()

            # Keep paragraph structure
            paragraphs = soup.find_all(['p', 'h1', 'h2', 'h3', 'li'])
            text = '\n\n'.join(p.get_text(strip=True) for p in paragraphs if p.get_text(strip=True))

            results[url] = f"{title}\n\n{text}"

        except Exception as e:
            results[url] = f"Error: {e}"

    return results
```

```python
import pandas as pd

df_pocket = pd.read_csv('pocket.csv', sep=";", encoding='utf-8')
allurls = df_pocket['url'].tolist()
alltitles = df_pocket['title'].tolist()
```

```python
urls = allurls[800:1000]  # Limit to first 10 URLs for testing
titles = alltitles[800:1000]  # Corresponding titles
```

```python

raw_texts = extract_reader_text(urls)

for url, content in raw_texts.items():
    print(f"\n--- Content from {url} ---\n")
    #title = content[:18]+".txt"
    #title = url.split("/")[-1] + ".txt"  # Use the last part of the URL as the filename
    title = titles[urls.index(url)] + ".txt"  # Use the corresponding title from the list
    title = title.replace("/", "_").replace("\\", "_")  # Replace slashes to avoid file path issues
    title = title.replace(":", "_").replace("?", "_")  # Replace colons and question marks
    title = title.replace("*", "_").replace("<", "_").replace(">", "_")  # Replace invalid filename characters
    title = title.replace('"', '_').replace('|', '_')  # Replace quotes and pipes
    title = title.replace(" ", "_")  # Replace spaces with underscores
    title = title[:100]  # Limit filename length to 100 characters
    title = title.strip()  # Remove leading/trailing whitespace
    title = title+ ".txt"  # Ensure it ends with .txt
    #title = title.replace(" ", "_")  # Replace spaces with underscores
    print(title)
    #print(content[:10000])  # Print first 1000 characters to keep it short
    with open("./files/"+title, 'w', encoding='utf-8') as f:
        f.write(raw_texts[url])
        f.write("\n\n--- End of Content ---\n")
        f.write(f"Source URL: {url}\n")
        f.close

```
